# GM Tensorflow Tutorial

```python
import tensorflow as tf
```

# Tensorflow Beginners guide on the mnist dataset

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
mnist
```

### Setting up the model parameters

```python
x = tf.placeholder(tf.float32, [None, 784])
```

```python
x
```

```python
W = tf.Variable(tf.zeros([784, 10]))
```

```python
b = tf.Variable(tf.zeros([10]))
```

```python
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

## Training the model

```python
y_ = tf.placeholder(tf.float32, [None, 10])
```

```python
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
sess = tf.InteractiveSession()
```

```python
tf.global_variables_initializer().run()
```

```python
for _ in range(1000):
  batch_xs, batch_ys = mnist.train.next_batch(100)
  sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

## Evaluating our model

```python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
```

## Saving the model

```python
export_dir = "my_model"
```

```python
builder = tf.saved_model.builder.SavedModelBuilder(export_dir)
```

```python
tag_constants = tf.saved_model.tag_constants
```

```python
builder.add_meta_graph_and_variables(sess,
                                       [tag_constants.TRAINING])
```

```python
builder.save()
```

## Loading saved model

```python
load_sess = tf.InteractiveSession()
```

```python
tf.saved_model.loader.load(load_sess, [tag_constants.TRAINING], export_dir)
```

### re-evaluating the loaded model to see if it performs just like before it was saved

```python
print(load_sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
```

### We can clearly see we got our saved model back
